```python
def decimal_to_binary(decimal_ip):
    octets = decimal_ip.split('.')
    binary_ip = ''
    for octet in octets:
        binary_octet = bin(int(octet))[2:].zfill(8) # zfill добавляет столько 0, сколько осталось свободных символов
        binary_ip += binary_octet + '.'
    binary_ip = binary_ip[:-1]  # Удаляем последнюю точку
    return binary_ip


def binary_to_decimal(binary_ip):
    octets = binary_ip.split('.')
    decimal_ip = ''
    for octet in octets:
        decimal_octet = str(int(octet, 2))
        decimal_ip += decimal_octet + '.'
    decimal_ip = decimal_ip[:-1]  # удаляем последнюю точку
    return decimal_ip

decimal_ip = "192.168.0.1"
binary_ip = decimal_to_binary(decimal_ip)
print(binary_ip)

binary_ip = "11000000.10101000.00000000.00000001"
decimal_ip = binary_to_decimal(binary_ip)
print(decimal_ip)
```

```python
def network_details(ip_ad):
    octets = ip_ad.split('.')
    first_octet = int(octets[0])

    if 0 <= first_octet <= 127:
        network_class = 'A'
        start_ip = f"{first_octet}.0.0.0"
        end_ip = f"{first_octet}.255.255.255"
        subnet_mask = "255.0.0.0"
    elif 128 <= first_octet <= 191:
        network_class = 'B'
        start_ip = f"{octets[0]}.{octets[1]}.0.0"
        end_ip = f"{octets[0]}.{octets[1]}.255.255"
        subnet_mask = "255.255.0.0"
    elif 192 <= first_octet <= 223:
        network_class = 'C'
        start_ip = f"{octets[0]}.{octets[1]}.{octets[2]}.0"
        end_ip = f"{octets[0]}.{octets[1]}.{octets[2]}.255"
        subnet_mask = "255.255.255.0"
    elif 224 <= first_octet <= 239:
        network_class = 'D'
        start_ip = ""
        end_ip = ""
        subnet_mask = ""
    elif 240 <= first_octet <= 255:
        network_class = 'E'
        start_ip = ""
        end_ip = ""
        subnet_mask = ""
    else:
        network_class = 'Класс сети не определён'
        start_ip = ""
        end_ip = ""
        subnet_mask = ""

    return network_class, start_ip, end_ip, subnet_mask
    

ip_ad = '192.168.0.1'
network_class, start_ip, end_ip, subnet_mask = network_details(ip_ad)

print("Класс сети:", network_class)
print("Начальный IP-адрес:", start_ip)
print("Конечный IP-адрес:", end_ip)
print("Маска подсети:", subnet_mask)
```

```python
import ipaddress


# Классы D и E не рассматриваются.
def net_class(ip_ad):
    octets = ip_ad.split('.')
    first_octet = int(octets[0])

    if 0 <= first_octet <= 127:
        network_class = 'A'
        N = 8
    elif 128 <= first_octet <= 191:
        network_class = 'B'
        N = 16
    elif 192 <= first_octet <= 223:
        network_class = 'C'
        N = 24
    else:
        print("Ошибка.")
        return

    return network_class, N


def get_details(ip_ad, num_subnets, num_hosts_per_subnet):
    network_class, N = net_class(ip_ad)

    # Определение количества бит для подсетей и хостов
    for i in range(1, 32):
        if 2 ** i >= num_subnets:
            S = i
            H = 32 - S - N
            break

    if 2 ** H - 2 >= num_hosts_per_subnet:
        # Создание маски подсети
        mask = str(ipaddress.IPv4Network('0.0.0.0/' + str(S + N)).netmask)
    else:
        mask = None

    # Подсчёт количества IP-адресов
    num_addresses = 2 ** H

    # Подсчёт доступных IP-адресов
    available_addresses = num_addresses - 2

    # Первые и последние 5 доступных IP
    network = ipaddress.ip_network(f"{ip_ad}/{S + N}")
    hosts = list(network.hosts())
    ip_addresses_first = hosts[:5]
    ip_addresses_last = hosts[-5:]

    return (f"Адрес сети: {network}",
            f"Класс сети: {network_class}",
            f"Маска: {mask}",
            f"Кол-во IP: {num_addresses}",
            f"Кол-во достурных IP: {available_addresses}",
            f"Первые 5 доступных IP: {ip_addresses_first}",
            f"Последние 5 доступных IP: {ip_addresses_last}")


# Пример использования:
network_address = "172.16.0.0"
num_subnets = 120
num_hosts_per_subnet = 180

get_details_ = get_details(network_address, num_subnets, num_hosts_per_subnet)

for row in get_details_:
    print(row)
```
